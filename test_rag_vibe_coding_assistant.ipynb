````python
# @title –ö–ª–∞—Å—Å –¥–ª—è –Ω–∞—à–µ–π RAG-—Å–∏—Å—Ç–µ–º—ã
class RAGSystem:
    def __init__(self, openai_api_key=None):
        # –ü—Ä–æ–≤–µ—Ä—è–µ–º –Ω–∞–ª–∏—á–∏–µ GPU –∏ –Ω–∞—Å—Ç—Ä–æ–π–∫–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
        device = "cpu"
        if USE_GPU:
            try:
                import torch
                device = "cuda" if torch.cuda.is_available() else "cpu"
                if device == "cuda":
                    logger.info("üöÄ –ù–ê–•–£–ô! –£ –ù–ê–° –ï–°–¢–¨ GPU! –ë—É–¥–µ–º —é–∑–∞—Ç—å –µ–≥–æ –¥–ª—è —ç–º–±–µ–¥–¥–∏–Ω–≥–æ–≤!")
                else:
                    logger.info("üò¢ GPU –Ω–µ –æ–±–Ω–∞—Ä—É–∂–µ–Ω, –∏—Å–ø–æ–ª—å–∑—É–µ–º CPU. –ë—É–¥–µ—Ç –º–µ–¥–ª–µ–Ω–Ω–µ–µ...")
            except:
                logger.info("PyTorch –Ω–µ –Ω–∞–π–¥–µ–Ω –∏–ª–∏ –æ—à–∏–±–∫–∞ –ø—Ä–∏ –ø—Ä–æ–≤–µ—Ä–∫–µ GPU. –ò—Å–ø–æ–ª—å–∑—É–µ–º CPU.")
        else:
            logger.info("üîÑ –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ GPU –æ—Ç–∫–ª—é—á–µ–Ω–æ –≤ –Ω–∞—Å—Ç—Ä–æ–π–∫–∞—Ö. –ò—Å–ø–æ–ª—å–∑—É–µ–º CPU.")

        # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è —ç–º–±–µ–¥–¥–∏–Ω–≥–æ–≤ —Å –≤—ã–±—Ä–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª—å—é
        logger.info(f"üß† –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è —ç–º–±–µ–¥–¥–∏–Ω–≥–æ–≤ E5-large –Ω–∞ {device.upper()}...")
        self.embeddings = HuggingFaceEmbeddings(
            model_name="intfloat/e5-large",  # –¢–æ–ø–æ–≤–∞—è –º–æ–¥–µ–ª—å!
            model_kwargs={
                'device': device,
                'token': HF_TOKEN
            },
            encode_kwargs={'normalize_embeddings': True}
        )

        # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è LLM
        logger.info("ü§ñ –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è —è–∑—ã–∫–æ–≤–æ–π –º–æ–¥–µ–ª–∏ OpenAI...")
        self.llm = ChatOpenAI(
            model_name="gpt-4o-mini",
            temperature=0.7,
            api_key=openai_api_key or OPENAI_API_KEY
        )

        # –•—Ä–∞–Ω–∏–ª–∏—â–µ –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤
        self.vectorstore = None

        # –§–ª–∞–≥ –∑–∞–≥—Ä—É–∑–∫–∏ –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤
        self.is_loaded = False

        # –°–æ–∑–¥–∞–Ω–∏–µ —Å–ø–ª–∏—Ç—Ç–µ—Ä–∞ –¥–ª—è –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤
        logger.info(f"‚úÇÔ∏è –ù–∞—Å—Ç—Ä–æ–π–∫–∞ —Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—è –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤ (—Ä–∞–∑–º–µ—Ä —á–∞–Ω–∫–∞: {CHUNK_SIZE}, –ø–µ—Ä–µ–∫—Ä—ã—Ç–∏–µ: {CHUNK_OVERLAP})...")
        self.text_splitter = RecursiveCharacterTextSplitter(
            chunk_size=CHUNK_SIZE,
            chunk_overlap=CHUNK_OVERLAP,
            length_function=len,
            separators=["\n\n", "\n", " ", ""]
        )

        # –°–æ–∑–¥–∞–Ω–∏–µ —à–∞–±–ª–æ–Ω–∞ –ø—Ä–æ–º–ø—Ç–∞ –¥–ª—è –∑–∞–ø—Ä–æ—Å–æ–≤
        self.prompt_template = """
        –¢—ã ‚Äî –¥–µ—Ä–∑–∫–∏–π –∞—Å—Å–∏—Å—Ç–µ–Ω—Ç –∏–∑ —Ç–µ–ª–µ–≥—Ä–∞–º-–∫–∞–Ω–∞–ª–∞ Vibe Coding AI, –∫–æ—Ç–æ—Ä—ã–π –æ—Ç–≤–µ—á–∞–µ—Ç –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã –ø–æ –∫–Ω–∏–≥–µ "–ì—Ä–æ–∫–∞–µ–º –∞–ª–≥–æ—Ä–∏—Ç–º—ã".
        –î–∞–∂–µ –µ—Å–ª–∏ –≤ –∫–æ–Ω—Ç–µ–∫—Å—Ç–µ –∫–∞–∂–µ—Ç—Å—è, —á—Ç–æ –Ω–µ—Ç –ø—Ä—è–º–æ–≥–æ –æ—Ç–≤–µ—Ç–∞ –Ω–∞ –≤–æ–ø—Ä–æ—Å, –ø–æ–ø—Ä–æ–±—É–π –Ω–∞–π—Ç–∏ —Å–≤—è–∑–∞–Ω–Ω—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –∏ –¥–∞—Ç—å –ø–æ–ª–µ–∑–Ω—ã–π –æ—Ç–≤–µ—Ç.
        –ò—Å–ø–æ–ª—å–∑—É–π —Å–≤–æ–∏ –∑–Ω–∞–Ω–∏—è –æ–± –∞–ª–≥–æ—Ä–∏—Ç–º–∞—Ö, –µ—Å–ª–∏ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –Ω–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ.

        –ö–æ–Ω—Ç–µ–∫—Å—Ç –∏–∑ –∫–Ω–∏–≥–∏:
        {context}

        –í–æ–ø—Ä–æ—Å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è: {question}

        –ï—Å–ª–∏ –≤ –∫–æ–Ω—Ç–µ–∫—Å—Ç–µ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ —Å–æ–≤—Å–µ–º –Ω–µ—Ç –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –ø–æ –≤–æ–ø—Ä–æ—Å—É, —Ç–æ–ª—å–∫–æ —Ç–æ–≥–¥–∞ —Å–∫–∞–∂–∏
        "–ë–ª—è—Ç—å, –≤ –¥–æ–∫—É–º–µ–Ω—Ç–∞—Ö –Ω–µ—Ç –∏–Ω—Ñ—ã –ø–æ —ç—Ç–æ–º—É –≤–æ–ø—Ä–æ—Å—É. –ó–∞–≥—Ä—É–∑–∏ —á—Ç–æ-—Ç–æ –ø–æ–ª—É—á—à–µ –∏–ª–∏ —Å–ø—Ä–æ—Å–∏ —á—Ç–æ-—Ç–æ –¥—Ä—É–≥–æ–µ."

        –¢–≤–æ–π —Å—Ç–∏–ª—å - –¥–µ—Ä–∑–∫–∏–π, —Å –ø—Ä–æ–≥—Ä–∞–º–º–∏—Å—Ç—Å–∫–∏–º —Å–ª–µ–Ω–≥–æ–º. –û—Ç–≤–µ—á–∞–π —è—Ä–∫–æ –∏ —ç–Ω–µ—Ä–≥–∏—á–Ω–æ.

        –û—Ç–≤–µ—Ç:
        """

        self.PROMPT = PromptTemplate(
            template=self.prompt_template,
            input_variables=["context", "question"]
        )

    async def load_document(self, file_path: str) -> List[Document]:
        """–ó–∞–≥—Ä—É–∑–∫–∞ –¥–æ–∫—É–º–µ–Ω—Ç–∞ –∏–∑ —Ñ–∞–π–ª–∞"""
        logger.info(f"üìÑ –ó–∞–≥—Ä—É–∑–∫–∞ –¥–æ–∫—É–º–µ–Ω—Ç–∞: {file_path}")

        try:
            if file_path.lower().endswith('.pdf'):
                loader = PyMuPDFLoader(file_path)
            elif file_path.lower().endswith('.txt'):
                loader = TextLoader(file_path)
            elif file_path.lower().endswith('.docx'):
                loader = UnstructuredFileLoader(file_path)
            else:
                raise ValueError(f"–ù–µ–ø–æ–¥–¥–µ—Ä–∂–∏–≤–∞–µ–º—ã–π —Ñ–æ—Ä–º–∞—Ç —Ñ–∞–π–ª–∞: {file_path}")

            # –í—ã–ø–æ–ª–Ω—è–µ–º –∑–∞–≥—Ä—É–∑–∫—É –≤ –æ—Ç–¥–µ–ª—å–Ω–æ–º –ø–æ—Ç–æ–∫–µ
            documents = await asyncio.to_thread(lambda: loader.load())

            logger.info(f"‚úÖ –ó–∞–≥—Ä—É–∂–µ–Ω–æ {len(documents)} —Å—Ç—Ä–∞–Ω–∏—Ü/—á–∞—Å—Ç–µ–π –∏–∑ {file_path}")
            return documents
        except Exception as e:
            logger.error(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –∑–∞–≥—Ä—É–∑–∫–µ –¥–æ–∫—É–º–µ–Ω—Ç–∞ {file_path}: {str(e)}")
            return []

    async def process_documents(self, documents: List[Document]) -> None:
        """–û–±—Ä–∞–±–æ—Ç–∫–∞ –∏ —Å–æ–∑–¥–∞–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–Ω–æ–≥–æ —Ö—Ä–∞–Ω–∏–ª–∏—â–∞ –∏–∑ –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤"""
        if not documents:
            logger.error("‚ùå –ù–µ—Ç –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤ –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏!")
            return

        logger.info("‚úÇÔ∏è –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤ –Ω–∞ —á–∞–Ω–∫–∏...")

        # –í—ã–ø–æ–ª–Ω—è–µ–º —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ –≤ –æ—Ç–¥–µ–ª—å–Ω–æ–º –ø–æ—Ç–æ–∫–µ
        chunks = await asyncio.to_thread(self.text_splitter.split_documents, documents)
        logger.info(f"‚úÖ –°–æ–∑–¥–∞–Ω–æ {len(chunks)} —á–∞–Ω–∫–æ–≤")

        logger.info("üß† –°–æ–∑–¥–∞–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–Ω–æ–≥–æ —Ö—Ä–∞–Ω–∏–ª–∏—â–∞...")

        # –°–æ–∑–¥–∞–µ–º –≤–µ–∫—Ç–æ—Ä–Ω–æ–µ —Ö—Ä–∞–Ω–∏–ª–∏—â–µ –≤ –æ—Ç–¥–µ–ª—å–Ω–æ–º –ø–æ—Ç–æ–∫–µ
        if self.vectorstore is None:
            self.vectorstore = await asyncio.to_thread(FAISS.from_documents, chunks, self.embeddings)
        else:
            # –î–æ–±–∞–≤–ª—è–µ–º –Ω–æ–≤—ã–µ –¥–æ–∫—É–º–µ–Ω—Ç—ã –∫ —Å—É—â–µ—Å—Ç–≤—É—é—â–µ–º—É —Ö—Ä–∞–Ω–∏–ª–∏—â—É
            await asyncio.to_thread(lambda: self.vectorstore.add_documents(chunks))

        logger.info("‚úÖ –í–µ–∫—Ç–æ—Ä–Ω–æ–µ —Ö—Ä–∞–Ω–∏–ª–∏—â–µ —Å–æ–∑–¥–∞–Ω–æ")
        self.is_loaded = True

    async def save_vectorstore(self, directory: str = VECTORSTORE_PATH) -> None:
        """–°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ –≤–µ–∫—Ç–æ—Ä–Ω–æ–≥–æ —Ö—Ä–∞–Ω–∏–ª–∏—â–∞ –Ω–∞ –¥–∏—Å–∫"""
        if not self.is_loaded:
            logger.warning("‚ö†Ô∏è –ù–µ—Ç –∑–∞–≥—Ä—É–∂–µ–Ω–Ω—ã—Ö –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤ –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è!")
            return

        # –°–æ–∑–¥–∞–µ–º –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—é, –µ—Å–ª–∏ –æ–Ω–∞ –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç
        os.makedirs(directory, exist_ok=True)

        await asyncio.to_thread(self.vectorstore.save_local, directory)
        logger.info(f"üíæ –í–µ–∫—Ç–æ—Ä–Ω–æ–µ —Ö—Ä–∞–Ω–∏–ª–∏—â–µ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–æ –≤ {directory}")

    async def load_vectorstore(self, directory: str = VECTORSTORE_PATH, allow_dangerous_deserialization: bool = False) -> None:
        """–ó–∞–≥—Ä—É–∑–∫–∞ –≤–µ–∫—Ç–æ—Ä–Ω–æ–≥–æ —Ö—Ä–∞–Ω–∏–ª–∏—â–∞ —Å –¥–∏—Å–∫–∞"""
        if not os.path.exists(directory):
            logger.warning(f"‚ö†Ô∏è –î–∏—Ä–µ–∫—Ç–æ—Ä–∏—è {directory} –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç!")
            return

        # –ò—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–∞—è –≤–µ—Ä—Å–∏—è —Å –ø–æ–¥–¥–µ—Ä–∂–∫–æ–π –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ allow_dangerous_deserialization
        self.vectorstore = await asyncio.to_thread(
            lambda: FAISS.load_local(
                directory,
                self.embeddings,
                allow_dangerous_deserialization=allow_dangerous_deserialization
            )
        )
        logger.info(f"üìÇ –í–µ–∫—Ç–æ—Ä–Ω–æ–µ —Ö—Ä–∞–Ω–∏–ª–∏—â–µ –∑–∞–≥—Ä—É–∂–µ–Ω–æ –∏–∑ {directory}")
        self.is_loaded = True

    async def load_documents_from_drive(self, folder_path: str) -> List[str]:
        """–ó–∞–≥—Ä—É–∑–∫–∞ –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤ –∏–∑ —É–∫–∞–∑–∞–Ω–Ω–æ–π –ø–∞–ø–∫–∏ –Ω–∞ Google Drive"""
        if not IN_COLAB and not os.path.exists(folder_path):
            logger.error(f"‚ùå –ü–∞–ø–∫–∞ –Ω–µ –Ω–∞–π–¥–µ–Ω–∞: {folder_path}")
            return []

        logger.info(f"üîç –ü–æ–∏—Å–∫ –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤ –≤ –ø–∞–ø–∫–µ {folder_path}...")

        # –ü–æ–ª—É—á–∞–µ–º —Å–ø–∏—Å–æ–∫ —Ñ–∞–π–ª–æ–≤ –≤ –ø–∞–ø–∫–µ
        try:
            files = []
            supported_extensions = ['.pdf', '.txt', '.docx']

            # –†–µ–∫—É—Ä—Å–∏–≤–Ω—ã–π –æ–±—Ö–æ–¥ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–π
            for root, _, filenames in os.walk(folder_path):
                for filename in filenames:
                    ext = os.path.splitext(filename)[1].lower()
                    if ext in supported_extensions:
                        files.append({
                            'name': filename,
                            'path': os.path.join(root, filename)
                        })

            logger.info(f"üìö –ù–∞–π–¥–µ–Ω–æ {len(files)} –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤ –≤ –ø–∞–ø–∫–µ {folder_path}")

            if not files:
                logger.warning("‚ö†Ô∏è –í —É–∫–∞–∑–∞–Ω–Ω–æ–π –ø–∞–ø–∫–µ –Ω–µ—Ç –ø–æ–¥–¥–µ—Ä–∂–∏–≤–∞–µ–º—ã—Ö –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤!")
                return []
        except Exception as e:
            logger.error(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –ø–æ–∏—Å–∫–µ –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤: {str(e)}")
            return []

        all_documents = []
        loaded_files = []

        # –ó–∞–≥—Ä—É–∂–∞–µ–º –∫–∞–∂–¥—ã–π —Ñ–∞–π–ª –∏ –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º –µ–≥–æ
        for file in files:
            try:
                logger.info(f"‚è¨ –ó–∞–≥—Ä—É–∑–∫–∞ —Ñ–∞–π–ª–∞ {file['name']}...")

                # –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–æ–∫—É–º–µ–Ω—Ç
                documents = await self.load_document(file['path'])

                if not documents:
                    continue

                all_documents.extend(documents)

                # –î–æ–±–∞–≤–ª—è–µ–º –∏–º—è —Ñ–∞–π–ª–∞ –≤ —Å–ø–∏—Å–æ–∫ –∑–∞–≥—Ä—É–∂–µ–Ω–Ω—ã—Ö
                loaded_files.append(file['name'])

            except Exception as e:
                logger.error(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –æ–±—Ä–∞–±–æ—Ç–∫–µ —Ñ–∞–π–ª–∞ {file['name']}: {str(e)}")

        # –û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º –≤—Å–µ –∑–∞–≥—Ä—É–∂–µ–Ω–Ω—ã–µ –¥–æ–∫—É–º–µ–Ω—Ç—ã
        if all_documents:
            logger.info(f"üß© –û–±—Ä–∞–±–æ—Ç–∫–∞ {len(all_documents)} –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤...")
            await self.process_documents(all_documents)
            return loaded_files
        else:
            logger.warning("‚ö†Ô∏è –ù–µ —É–¥–∞–ª–æ—Å—å –∑–∞–≥—Ä—É–∑–∏—Ç—å –Ω–∏ –æ–¥–Ω–æ–≥–æ –¥–æ–∫—É–º–µ–Ω—Ç–∞!")
            return []

    async def query(self, question: str) -> str:
        """–ó–∞–ø—Ä–æ—Å –∫ —Å–∏—Å—Ç–µ–º–µ"""
        if not self.is_loaded:
            return "‚ö†Ô∏è –°–Ω–∞—á–∞–ª–∞ –∑–∞–≥—Ä—É–∑–∏ –¥–æ–∫—É–º–µ–Ω—Ç—ã! –ö–∞–∫–æ–≥–æ —Ö—É—è —è –¥–æ–ª–∂–µ–Ω –æ—Ç–≤–µ—á–∞—Ç—å –Ω–∞ –ø—É—Å—Ç–æ—Ç—É?!"

        try:
            start_time = time.time()

            # –ü–æ–ª—É—á–∞–µ–º —Ä–µ–ª–µ–≤–∞–Ω—Ç–Ω—ã–µ —á–∞–Ω–∫–∏
            retriever = self.vectorstore.as_retriever(search_kwargs={"k": 10})

            # –°–æ–∑–¥–∞–µ–º —Ü–µ–ø–æ—á–∫—É –¥–ª—è –æ—Ç–≤–µ—Ç–∞ –Ω–∞ –≤–æ–ø—Ä–æ—Å
            qa_chain = RetrievalQA.from_chain_type(
                llm=self.llm,
                chain_type="stuff",
                retriever=retriever,
                return_source_documents=True,
                chain_type_kwargs={"prompt": self.PROMPT}
            )

            # –í—ã–ø–æ–ª–Ω—è–µ–º –∑–∞–ø—Ä–æ—Å –≤ –æ—Ç–¥–µ–ª—å–Ω–æ–º –ø–æ—Ç–æ–∫–µ
            result = await asyncio.to_thread(lambda: qa_chain({"query": question}))

            # –§–æ—Ä–º–∞—Ç–∏—Ä—É–µ–º –æ—Ç–≤–µ—Ç —Å –∏—Å—Ç–æ—á–Ω–∏–∫–∞–º–∏
            answer = result["result"]
            sources = set()

            # –î–æ–±–∞–≤–ª—è–µ–º —Å–ø–∏—Å–æ–∫ –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤
            for doc in result["source_documents"]:
                if hasattr(doc, 'metadata') and 'source' in doc.metadata:
                    source_path = doc.metadata['source']
                    source_name = os.path.basename(source_path)
                    sources.add(source_name)

            elapsed_time = time.time() - start_time
            formatted_answer = f"{answer}\n\n"

            if sources:
                formatted_answer += f"üìö –ò—Å—Ç–æ—á–Ω–∏–∫–∏: {', '.join(sources)}\n"

            formatted_answer += f"‚è±Ô∏è –í—Ä–µ–º—è –æ—Ç–≤–µ—Ç–∞: {elapsed_time:.2f} —Å–µ–∫"

            return formatted_answer

        except Exception as e:
            logger.error(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–∏ –∑–∞–ø—Ä–æ—Å–∞: {str(e)}")
            return f"–ü—Ä–æ–∏–∑–æ—à–ª–∞ –æ—à–∏–±–∫–∞ –ø—Ä–∏ –æ–±—Ä–∞–±–æ—Ç–∫–µ –∑–∞–ø—Ä–æ—Å–∞: {str(e)}"
````

````python
# @title üìä –ú–æ–Ω–∏—Ç–æ—Ä–∏–Ω–≥ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ GPU
# @markdown –ü—Ä–æ–≤–µ—Ä—å—Ç–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ GPU –ø–µ—Ä–µ–¥ –∑–∞–ø—É—Å–∫–æ–º —Å–∏—Å—Ç–µ–º—ã:

def check_gpu():
    """–ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞–ª–∏—á–∏—è –∏ —Å–æ—Å—Ç–æ—è–Ω–∏—è GPU"""
    try:
        import torch
        if torch.cuda.is_available():
            gpu_info = {
                "–î–æ—Å—Ç—É–ø–µ–Ω": "–î–∞",
                "–ù–∞–∑–≤–∞–Ω–∏–µ": torch.cuda.get_device_name(0),
                "–ü–∞–º—è—Ç—å –≤—Å–µ–≥–æ (GB)": round(torch.cuda.get_device_properties(0).total_memory / (1024**3), 2),
                "–ü–∞–º—è—Ç—å –¥–æ—Å—Ç—É–ø–Ω–æ (GB)": round(torch.cuda.memory_reserved(0) / (1024**3), 2)
            }

            if IN_COLAB:
                from IPython.display import HTML, display

                # –°–æ–∑–¥–∞–µ–º —Ç–∞–±–ª–∏—Ü—É –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è
                table_html = "<table style='width:100%; border-collapse:collapse; margin-top:10px;'>"
                table_html += "<tr style='background-color:#4CAF50; color:white;'><th style='padding:8px; text-align:left;'>–ü–∞—Ä–∞–º–µ—Ç—Ä</th><th style='padding:8px; text-align:left;'>–ó–Ω–∞—á–µ–Ω–∏–µ</th></tr>"

                for i, (key, value) in enumerate(gpu_info.items()):
                    bg_color = "#f2f2f2" if i % 2 == 0 else "white"
                    table_html += f"<tr style='background-color:{bg_color};'><td style='padding:8px; border:1px solid #ddd;'>{key}</td><td style='padding:8px; border:1px solid #ddd;'>{value}</td></tr>"

                table_html += "</table>"

                display(HTML(f"""
                <div style="background-color:#d4edda; padding:20px; border-radius:5px; color:#155724; margin-bottom:20px">
                    <h3 style="margin-top:0">‚úÖ GPU –¥–æ—Å—Ç—É–ø–µ–Ω!</h3>
                    {table_html}
                </div>
                """))
            else:
                print("‚úÖ GPU –¥–æ—Å—Ç—É–ø–µ–Ω!")
                for key, value in gpu_info.items():
                    print(f"{key}: {value}")
        else:
            if IN_COLAB:
                from IPython.display import HTML, display
                display(HTML("""
                <div style="background-color:#f8d7da; padding:20px; border-radius:5px; color:#721c24; margin-bottom:20px">
                    <h3 style="margin-top:0">‚ùå GPU –Ω–µ –¥–æ—Å—Ç—É–ø–µ–Ω!</h3>
                    <p>–ü—Ä–æ–≤–µ—Ä—å—Ç–µ –Ω–∞—Å—Ç—Ä–æ–π–∫–∏ —Å—Ä–µ–¥—ã –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è Colab (Runtime > Change runtime type > Hardware accelerator > GPU)</p>
                </div>
                """))
            else:
                print("‚ùå GPU –Ω–µ –¥–æ—Å—Ç—É–ø–µ–Ω!")
    except Exception as e:
        if IN_COLAB:
            from IPython.display import HTML, display
            display(HTML(f"""
            <div style="background-color:#f8d7da; padding:20px; border-radius:5px; color:#721c24; margin-bottom:20px">
                <h3 style="margin-top:0">‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –ø—Ä–æ–≤–µ—Ä–∫–µ GPU!</h3>
                <p>–û—à–∏–±–∫–∞: {str(e)}</p>
            </div>
            """))
        else:
            print(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –ø—Ä–æ–≤–µ—Ä–∫–µ GPU: {str(e)}")

# –ó–∞–ø—É—Å–∫–∞–µ–º –ø—Ä–æ–≤–µ—Ä–∫—É GPU
check_gpu()
````

````python
# –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è —Ä–∞–±–æ—Ç—ã RAG-—Å–∏—Å—Ç–µ–º—ã –≤ Colab
# @title üöÄ –°–æ–∑–¥–∞–Ω–∏–µ –∏ —Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ RAG-—Å–∏—Å—Ç–µ–º—ã
# @markdown –ù–∞–∂–º–∏—Ç–µ –∫–Ω–æ–ø–∫—É –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è RAG-—Å–∏—Å—Ç–µ–º—ã:

if IN_COLAB:
    from IPython.display import HTML, display
    display(HTML("""
    <div style="background-color:#d4edda; padding:20px; border-radius:5px; color:#155724; margin-bottom:20px">
        <h3 style="margin-top:0">üöÄ –°–æ–∑–¥–∞–µ–º –∏ —Ç–µ—Å—Ç–∏—Ä—É–µ–º RAG-—Å–∏—Å—Ç–µ–º—É</h3>
        <p>–°–µ–π—á–∞—Å –º—ã –∑–∞–≥—Ä—É–∑–∏–º –¥–æ–∫—É–º–µ–Ω—Ç—ã —Å Google Drive –∏ —Å–æ–∑–¥–∞–¥–∏–º –±–∞–∑—É –∑–Ω–∞–Ω–∏–π.</p>
    </div>
    """))

# –°–æ–∑–¥–∞–µ–º –≥–ª–æ–±–∞–ª—å–Ω—É—é RAG-—Å–∏—Å—Ç–µ–º—É
rag_system = RAGSystem()

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∑–∞–≥—Ä—É–∑–∫–∏ –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤ —Å Google Drive
async def load_docs_from_drive():
    """–ó–∞–≥—Ä—É–∑–∫–∞ –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤ —Å Google Drive"""
    loaded_files = await rag_system.load_documents_from_drive(DOCS_FOLDER_PATH)

    if loaded_files:
        # –°–æ—Ö—Ä–∞–Ω—è–µ–º –≤–µ–∫—Ç–æ—Ä–Ω–æ–µ —Ö—Ä–∞–Ω–∏–ª–∏—â–µ
        await rag_system.save_vectorstore()

        if IN_COLAB:
            from IPython.display import HTML, display
            file_list = "".join([f"<li>{file}</li>" for file in loaded_files])
            display(HTML(f"""
            <div style="background-color:#d4edda; padding:20px; border-radius:5px; color:#155724">
                <h3 style="margin-top:0">‚úÖ –£—Å–ø–µ—à–Ω–æ –∑–∞–≥—Ä—É–∂–µ–Ω–æ {len(loaded_files)} –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤!</h3>
                <p>–ó–∞–≥—Ä—É–∂–µ–Ω–Ω—ã–µ —Ñ–∞–π–ª—ã:</p>
                <ul>{file_list}</ul>
            </div>
            """))
    else:
        if IN_COLAB:
            from IPython.display import HTML, display
            display(HTML("""
            <div style="background-color:#f8d7da; padding:20px; border-radius:5px; color:#721c24">
                <h3 style="margin-top:0">‚ùå –ù–µ —É–¥–∞–ª–æ—Å—å –∑–∞–≥—Ä—É–∑–∏—Ç—å –¥–æ–∫—É–º–µ–Ω—Ç—ã!</h3>
                <p>–ü—Ä–æ–≤–µ—Ä—å—Ç–µ –ø—É—Ç—å –∫ –ø–∞–ø–∫–µ –∏ –Ω–∞–ª–∏—á–∏–µ –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤.</p>
            </div>
            """))

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∑–∞–≥—Ä—É–∑–∫–∏ –≤–µ–∫—Ç–æ—Ä–Ω–æ–≥–æ —Ö—Ä–∞–Ω–∏–ª–∏—â–∞
async def load_vectorstore():
    """–ó–∞–≥—Ä—É–∑–∫–∞ –≤–µ–∫—Ç–æ—Ä–Ω–æ–≥–æ —Ö—Ä–∞–Ω–∏–ª–∏—â–∞"""
    try:
        # –î–æ–±–∞–≤–ª—è–µ–º —Ñ–ª–∞–≥ –¥–ª—è –±–µ–∑–æ–ø–∞—Å–Ω–æ–π –¥–µ—Å–µ—Ä–∏–∞–ª–∏–∑–∞—Ü–∏–∏
        await rag_system.load_vectorstore(allow_dangerous_deserialization=True)

        if rag_system.is_loaded:
            if IN_COLAB:
                from IPython.display import HTML, display
                display(HTML("""
                <div style="background-color:#d4edda; padding:20px; border-radius:5px; color:#155724">
                    <h3 style="margin-top:0">‚úÖ –í–µ–∫—Ç–æ—Ä–Ω–æ–µ —Ö—Ä–∞–Ω–∏–ª–∏—â–µ —É—Å–ø–µ—à–Ω–æ –∑–∞–≥—Ä—É–∂–µ–Ω–æ!</h3>
                    <p>–¢–µ–ø–µ—Ä—å –º–æ–∂–Ω–æ –∑–∞–¥–∞–≤–∞—Ç—å –≤–æ–ø—Ä–æ—Å—ã.</p>
                </div>
                """))
        else:
            if IN_COLAB:
                from IPython.display import HTML, display
                display(HTML("""
                <div style="background-color:#f8d7da; padding:20px; border-radius:5px; color:#721c24">
                    <h3 style="margin-top:0">‚ùå –ù–µ —É–¥–∞–ª–æ—Å—å –∑–∞–≥—Ä—É–∑–∏—Ç—å –≤–µ–∫—Ç–æ—Ä–Ω–æ–µ —Ö—Ä–∞–Ω–∏–ª–∏—â–µ!</h3>
                    <p>–ü—Ä–æ–≤–µ—Ä—å—Ç–µ –ø—É—Ç—å –∫ —Ö—Ä–∞–Ω–∏–ª–∏—â—É –∏ –Ω–∞–ª–∏—á–∏–µ —Ñ–∞–π–ª–æ–≤.</p>
                </div>
                """))
    except Exception as e:
        if IN_COLAB:
            from IPython.display import HTML, display
            display(HTML(f"""
            <div style="background-color:#f8d7da; padding:20px; border-radius:5px; color:#721c24">
                <h3 style="margin-top:0">‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –∑–∞–≥—Ä—É–∑–∫–µ –≤–µ–∫—Ç–æ—Ä–Ω–æ–≥–æ —Ö—Ä–∞–Ω–∏–ª–∏—â–∞!</h3>
                <p>–û—à–∏–±–∫–∞: {str(e)}</p>
            </div>
            """))

async def test_query(question):
    """–¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –∑–∞–ø—Ä–æ—Å–∞ –∫ RAG-—Å–∏—Å—Ç–µ–º–µ"""
    if not rag_system.is_loaded:
        if IN_COLAB:
            from IPython.display import HTML, display
            display(HTML("""
            <div style="background-color:#FF4136; padding:20px; border-radius:10px; color:white; box-shadow: 0 4px 8px rgba(0,0,0,0.2);">
                <h3 style="margin-top:0; font-size:24px;">‚õîÔ∏è –°–¢–û–ü, –ù–ê–•–†–ï–ù!</h3>
                <p style="font-size:18px;">–ë–∞–∑–∞ –∑–Ω–∞–Ω–∏–π –Ω–µ –∑–∞–≥—Ä—É–∂–µ–Ω–∞! –ö–∞–∫–æ–≥–æ —Ö—Ä–µ–Ω–∞ —Ç—ã –ø—ã—Ç–∞–µ—à—å—Å—è –∑–∞–ø—Ä–æ—Å–∏—Ç—å?</p>
                <p style="font-size:16px; margin-top:10px;">üëâ –°–Ω–∞—á–∞–ª–∞ –∑–∞–≥—Ä—É–∑–∏ –¥–æ–∫—É–º–µ–Ω—Ç—ã, –ø–æ—Ç–æ–º –ø–æ–≥–æ–≤–æ—Ä–∏–º!</p>
            </div>
            """))
        return

    answer = await rag_system.query(question)
    answer_html = answer.replace('\n', '<br>')

    if IN_COLAB:
        from IPython.display import HTML, display
        display(HTML(f"""
        <div style="background-color:#111; color:#fff; padding:25px; border-radius:15px; margin-top:20px; box-shadow: 0 6px 12px rgba(0,0,0,0.3); border-left: 5px solid #39CCCC;">
            <h3 style="margin-top:0; color:#FF851B; font-size:24px; text-transform:uppercase;">üî• –í–æ–ø—Ä–æ—Å:</h3>
            <p style="font-size:18px; background-color:#222; padding:15px; border-radius:8px;">{question}</p>

            <h3 style="color:#2ECC40; font-size:24px; margin-top:20px; text-transform:uppercase;">üí• –û—Ç–≤–µ—Ç:</h3>
            <div style="font-size:18px; background-color:#222; padding:15px; border-radius:8px; line-height:1.6;">{answer_html}</div>

            <div style="margin-top:15px; font-size:14px; color:#999; text-align:right;">Powered by Vibe Coding AI ü§ò</div>
        </div>
        """))
    else:
        print("\n" + "=" * 50)
        print(f"üî• –í–û–ü–†–û–°: {question}")
        print("=" * 50)
        print(f"üí• –û–¢–í–ï–¢: {answer}")
        print("=" * 50 + "\n")
````

````python
# @title –¢–µ–ø–µ—Ä—å –∏–Ω—Ç–µ–≥—Ä–∏—Ä—É–µ–º Telegram-–±–æ—Ç–∞ –¥–ª—è –¥–æ—Å—Ç—É–ø–∞ –∫ RAG-—Å–∏—Å—Ç–µ–º–µ

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –±–æ—Ç–∞
bot = Bot(token=TELEGRAM_TOKEN)
dp = Dispatcher(storage=MemoryStorage())
router = Router()

# –•—Ä–∞–Ω–∏–ª–∏—â–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏—Ö –¥–∞–Ω–Ω—ã—Ö
user_data: Dict[int, Dict[str, Any]] = {}

# –ö–æ–º–∞–Ω–¥–∞ /start
@router.message(CommandStart())
async def cmd_start(message: Message):
    """–û–±—Ä–∞–±–æ—Ç—á–∏–∫ –∫–æ–º–∞–Ω–¥—ã /start"""
    user_id = message.from_user.id

    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è, –µ—Å–ª–∏ –∏—Ö –Ω–µ—Ç
    if user_id not in user_data:
        user_data[user_id] = {
            "documents": []
        }

    # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –∑–∞–≥—Ä—É–∂–µ–Ω–∞ –ª–∏ –≥–ª–æ–±–∞–ª—å–Ω–∞—è RAG-—Å–∏—Å—Ç–µ–º–∞
    if not rag_system.is_loaded:
        await message.answer(
            "‚ö†Ô∏è RAG-—Å–∏—Å—Ç–µ–º–∞ –µ—â–µ –Ω–µ –∑–∞–≥—Ä—É–∂–µ–Ω–∞. –î–ª—è –Ω–∞—á–∞–ª–∞ —Ä–∞–±–æ—Ç—ã —Å –±–æ—Ç–æ–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ "
            "–∑–∞–≥—Ä—É–∑–∏—Ç—å –¥–æ–∫—É–º–µ–Ω—Ç—ã –≤ Colab-–Ω–æ—É—Ç–±—É–∫–µ –∏ –∑–∞–ø—É—Å—Ç–∏—Ç—å –±–æ—Ç–∞ –æ—Ç—Ç—É–¥–∞."
        )

    # –ü—Ä–∏–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ–µ —Å–æ–æ–±—â–µ–Ω–∏–µ
    start_text = Text(
        "üî• ", Bold("VIBE CODING AI"), " üî•\n\n",
        "–ü—Ä–∏–≤–µ—Ç! –Ø –ª–∏—á–Ω—ã–π RAG-–∞—Å—Å–∏—Å—Ç–µ–Ω—Ç —Å –Ω–æ—É—Ç–±—É–∫–æ–º –≤ Google Colab.\n\n",
        "–ß—Ç–æ —è –º–æ–≥—É:\n",
        "‚Ä¢ –û—Ç–≤–µ—á–∞—Ç—å –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã –ø–æ —Ç–≤–æ–µ–π –±–∞–∑–µ –∑–Ω–∞–Ω–∏–π\n",
        "‚Ä¢ –†–∞–±–æ—Ç–∞—Ç—å —Å –¥–æ–∫—É–º–µ–Ω—Ç–∞–º–∏, –∑–∞–≥—Ä—É–∂–µ–Ω–Ω—ã–º–∏ –≤ Colab\n",
        "‚Ä¢ –î–∞–≤–∞—Ç—å —Ç–æ—á–Ω—ã–µ –æ—Ç–≤–µ—Ç—ã –Ω–∞ –æ—Å–Ω–æ–≤–µ —Ç–≤–æ–∏—Ö –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤\n\n",
        "–ò—Å–ø–æ–ª—å–∑—É–π /help –¥–ª—è –ø–æ–¥—Ä–æ–±–Ω–æ–π –∏–Ω—Å—Ç—Ä—É–∫—Ü–∏–∏."
    )

    await message.answer(**start_text.as_kwargs())

# –ö–æ–º–∞–Ω–¥–∞ /help
@router.message(Command("help"))
async def cmd_help(message: Message):
    """–û–±—Ä–∞–±–æ—Ç—á–∏–∫ –∫–æ–º–∞–Ω–¥—ã /help"""
    help_text = Text(
        "üî• ", Bold("–î–û–°–¢–£–ü–ù–´–ï –ö–û–ú–ê–ù–î–´"), " üî•\n\n",
        "/start - –Ω–∞—á–∞—Ç—å —Ä–∞–±–æ—Ç—É —Å –±–æ—Ç–æ–º\n",
        "/status - –ø—Ä–æ–≤–µ—Ä–∏—Ç—å —Å—Ç–∞—Ç—É—Å —Å–∏—Å—Ç–µ–º—ã\n\n",
        "–ß—Ç–æ–±—ã –∑–∞–¥–∞—Ç—å –≤–æ–ø—Ä–æ—Å –∫ —Å–≤–æ–µ–π –±–∞–∑–µ –∑–Ω–∞–Ω–∏–π, –ø—Ä–æ—Å—Ç–æ –Ω–∞–ø–∏—à–∏ –µ–≥–æ –≤ —á–∞—Ç."
    )

    await message.answer(**help_text.as_kwargs())

# –ö–æ–º–∞–Ω–¥–∞ /status
@router.message(Command("status"))
async def cmd_status(message: Message):
    """–û–±—Ä–∞–±–æ—Ç—á–∏–∫ –∫–æ–º–∞–Ω–¥—ã /status - –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —Ç–µ–∫—É—â–µ–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ —Å–∏—Å—Ç–µ–º—ã"""
    user_id = message.from_user.id

    if user_id not in user_data:
        user_data[user_id] = {
            "documents": []
        }

    # –°—Ç–∞—Ç—É—Å —Å–∏—Å—Ç–µ–º—ã
    status_text = Text(
        "üìä ", Bold("–°–¢–ê–¢–£–° –°–ò–°–¢–ï–ú–´"), "\n\n",
        f"–ë–∞–∑–∞ –∑–Ω–∞–Ω–∏–π –∑–∞–≥—Ä—É–∂–µ–Ω–∞: {'‚úÖ –î–∞' if rag_system.is_loaded else '‚ùå –ù–µ—Ç'}\n"
    )

    # –ï—Å–ª–∏ —É –Ω–∞—Å –µ—Å—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –∑–∞–≥—Ä—É–∂–µ–Ω–Ω—ã—Ö –¥–æ–∫—É–º–µ–Ω—Ç–∞—Ö, –¥–æ–±–∞–≤–ª—è–µ–º –µ–µ
    documents = user_data[user_id].get("documents", [])
    if documents:
        status_text += Text("üìö ", Bold("–ü–û–°–õ–ï–î–ù–ò–ï –ó–ê–ü–†–û–°–´"), ":\n")
        for i, doc in enumerate(documents[-5:], 1):
            status_text += Text(f"{i}. {doc}\n")
    else:
        status_text += Text("‚ùì –ï—â–µ –Ω–µ –±—ã–ª–æ –∑–∞–ø—Ä–æ—Å–æ–≤. –ó–∞–¥–∞–π—Ç–µ –≤–æ–ø—Ä–æ—Å!")

    await message.answer(**status_text.as_kwargs())

# –û–±—Ä–∞–±–æ—Ç—á–∏–∫ —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö —Å–æ–æ–±—â–µ–Ω–∏–π (–≤–æ–ø—Ä–æ—Å–æ–≤)
@router.message(F.text)
async def handle_message(message: Message):
    """–û–±—Ä–∞–±–æ—Ç—á–∏–∫ —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö —Å–æ–æ–±—â–µ–Ω–∏–π (–≤–æ–ø—Ä–æ—Å–æ–≤)"""
    user_id = message.from_user.id

    # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω—ã –ª–∏ –¥–∞–Ω–Ω—ã–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
    if user_id not in user_data:
        user_data[user_id] = {
            "documents": []
        }

    # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –∑–∞–≥—Ä—É–∂–µ–Ω—ã –ª–∏ –¥–æ–∫—É–º–µ–Ω—Ç—ã
    if not rag_system.is_loaded:
        await message.answer(
            "‚ö†Ô∏è –ë–∞–∑–∞ –∑–Ω–∞–Ω–∏–π –Ω–µ –∑–∞–≥—Ä—É–∂–µ–Ω–∞! –î–ª—è –Ω–∞—á–∞–ª–∞ —Ä–∞–±–æ—Ç—ã —Å –±–æ—Ç–æ–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ "
            "–∑–∞–≥—Ä—É–∑–∏—Ç—å –¥–æ–∫—É–º–µ–Ω—Ç—ã –≤ Colab-–Ω–æ—É—Ç–±—É–∫–µ –∏ –∑–∞–ø—É—Å—Ç–∏—Ç—å –±–æ—Ç–∞ –æ—Ç—Ç—É–¥–∞."
        )
        return

    # –ü–æ–ª—É—á–∞–µ–º –≤–æ–ø—Ä–æ—Å
    question = message.text

    # –°–æ—Ö—Ä–∞–Ω—è–µ–º –≤–æ–ø—Ä–æ—Å –≤ –∏—Å—Ç–æ—Ä–∏–∏
    user_data[user_id]["documents"].append(question)

    # –°–æ–æ–±—â–∞–µ–º –æ –Ω–∞—á–∞–ª–µ –æ–±—Ä–∞–±–æ—Ç–∫–∏
    processing_msg = await message.answer("üß† –û–±—Ä–∞–±–∞—Ç—ã–≤–∞—é –≤–æ–ø—Ä–æ—Å... –°–µ–∫—É–Ω–¥—É...")

    try:
        # –ü–æ–ª—É—á–∞–µ–º –æ—Ç–≤–µ—Ç –Ω–∞ –≤–æ–ø—Ä–æ—Å
        answer = await rag_system.query(question)

        # –û—Ç–ø—Ä–∞–≤–ª—è–µ–º –æ—Ç–≤–µ—Ç
        await message.answer(answer)
    except Exception as e:
        # –í —Å–ª—É—á–∞–µ –æ—à–∏–±–∫–∏
        await message.answer(f"‚ùå –û—à–∏–±–∫–∞ –ø—Ä–∏ –æ–±—Ä–∞–±–æ—Ç–∫–µ –≤–æ–ø—Ä–æ—Å–∞: {str(e)}")
    finally:
        # –£–¥–∞–ª—è–µ–º —Å–æ–æ–±—â–µ–Ω–∏–µ –æ –∑–∞–≥—Ä—É–∑–∫–µ
        try:
            await bot.delete_message(chat_id=message.chat.id, message_id=processing_msg.message_id)
        except aiogram.exceptions.TelegramBadRequest:
            # –ï—Å–ª–∏ —Å–æ–æ–±—â–µ–Ω–∏–µ —É–∂–µ —É–¥–∞–ª–µ–Ω–æ –∏–ª–∏ –ø—Ä–æ—à–ª–æ —Å–ª–∏—à–∫–æ–º –º–Ω–æ–≥–æ –≤—Ä–µ–º–µ–Ω–∏
            pass

# –§—É–Ω–∫—Ü–∏—è –∑–∞–ø—É—Å–∫–∞ –±–æ—Ç–∞
async def start_bot():
    """–ó–∞–ø—É—Å–∫ –±–æ—Ç–∞"""
    # –†–µ–≥–∏—Å—Ç—Ä–∏—Ä—É–µ–º —Ä–æ—É—Ç–µ—Ä
    dp.include_router(router)

    # –ü—Ä–æ–ø—É—Å–∫–∞–µ–º –Ω–∞–∫–æ–ø–∏–≤—à–∏–µ—Å—è –æ–±–Ω–æ–≤–ª–µ–Ω–∏—è –∏ –∑–∞–ø—É—Å–∫–∞–µ–º –±–æ—Ç–∞
    await bot.delete_webhook(drop_pending_updates=True)

    logger.info("üöÄ –ë–æ—Ç –∑–∞–ø—É—â–µ–Ω! VIBE CODING RAG-BOT –í –î–ï–õ–ï!")

    # –ó–∞–ø—É—Å–∫–∞–µ–º –±–æ—Ç–∞
    await dp.start_polling(bot, allowed_updates=dp.resolve_used_update_types())
````
